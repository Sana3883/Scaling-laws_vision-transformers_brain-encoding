```python
import matplotlib.pyplot as plt
import numpy as np

train_loss_10k = np.load("10K_16bit_batchsize16_val_loss_values.npy")
train_loss_100k = np.load("100K_16bit_batchsize16_val_loss_values.npy")
train_loss_1M = np.load("1M_16bit_batchsize16_val_loss_values.npy")
train_loss_6M = np.load("6M_16bit_batchsize16_val_loss_values.npy")

plt.plot(train_loss_10k, 'g*', label='10K')
plt.plot(train_loss_100k, 'bo', label='100K')
plt.plot(train_loss_1M, 'mo', label='1M')
plt.plot(train_loss_6M, 'ro', label='6M')


plt.xlabel('Epochs')
plt.title('Training Loss (Cross-Entropy)')
plt.legend()


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


sub1_loss = [0.395, 0.586, 0.614, 0.619]
sub2_loss = [0.464, 0.607, 0.65, 0.673]


colors = {'10K': 'g', '100K': 'b', '1M': 'm', '6M': 'r'}


datasets = ['10K', '100K', '1M', '6M']

bar_width = 0.3

plt.figure(figsize=(10, 5))
for i, dataset in enumerate(datasets):
    
    plt.bar(i - bar_width/2 - 0.03, sub1_loss[i], color=colors[dataset], width=bar_width, label='Sub-01 - ' + dataset)
  
    plt.bar(i + bar_width/2 + 0.03, sub2_loss[i] if i < len(sub2_loss) else 0, color=colors[dataset], hatch='/', width=bar_width, label='Sub-04 - ' + dataset, edgecolor='black')


plt.xticks(np.arange(len(datasets)), datasets)

plt.xlabel('Training Dataset')
plt.ylabel('Pearson Correlation Coefficient (r)')
plt.title('brain encoding correlation for "attn_stack.attn_nets.4.post_fc_dp" layer')


plt.legend(bbox_to_anchor=(1.04, 0.5), loc="center left")

plt.tight_layout()
plt.show()
```
