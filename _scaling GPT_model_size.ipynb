```python
import matplotlib.pyplot as plt
import datetime
import numpy as np


train_loss_6M_dim_63 = np.load("hidden_dim_63_val_loss_values.npy")
train_loss_6M_dim_126 = np.load("hidden_dim_126_val_loss_values.npy")
train_loss_6M_dim_255 = np.load("6M_hidden_dim_255_val_loss_values.npy")
train_loss_6M = np.load("6M_16bit_batchsize16_val_loss_values.npy")
train_loss_6M_dim_1024 = np.load("6M_hidden_dim_1024_16bit.npy")


plt.figure(figsize=(12, 8)) 
plt.plot(train_loss_6M_dim_63, 'go', label='hid_dim_63')
plt.plot(train_loss_6M_dim_126, 'bo', label='hid_dim_126')
plt.plot(train_loss_6M_dim_255, 'mo', label='hid_dim_255')
plt.plot(train_loss_6M , 'ko', label='hid_dim_576')
plt.plot(train_loss_6M_dim_1024, 'ro', label='hid_dim_1024')
plt.xlabel('Epochs',fontsize=24)
plt.ylabel('Cross-Entropy',fontsize=24)
plt.title('Training loss with various hidden dimensions in the GPT model',fontsize=24)
plt.legend(fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=24)
plt.show()

```
